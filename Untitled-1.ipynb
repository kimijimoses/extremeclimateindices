```python
import os
import pandas as pd


# Specify the folder path containing the CSV files
folder_path = "C:\\Users\\USER\\Desktop\\Winter2024\\climate change\\Assignment 3\\data\\exceldata"

# Initialize an empty list to store DataFrames
dfs = []

# Iterate over each file in the folder
for file in os.listdir(folder_path):
    # Check if the file is a CSV file
    if file.endswith(".csv"):
        # Read the CSV file into a DataFrame and append it to the list
        df = pd.read_csv(os.path.join(folder_path, file))
        dfs.append(df)

# Concatenate the list of DataFrames into a single DataFrame
combined_df = pd.concat(dfs, ignore_index=True)

# Display the combined DataFrame
print(combined_df)
```

```python
print(combined_df .columns.tolist())
```

```python
# List of column names to extract
columns_to_extract = ['Max Temp (°C)', 'Min Temp (°C)', 'Total Precip (mm)', 'Station Name', 'Date/Time', 'Year', 'Month', 'Day']

# Extract the specified columns
extracted_df = combined_df[columns_to_extract]

# Display the extracted DataFrame
print(extracted_df)
```

```python
# Filter out rows where 'Year' column has a value of 2012
filtered_df = extracted_df[extracted_df['Year'] != 2012]

filtered_df.tail()
```

```python
# Find the number of NaN values in the DataFrame
nan_count = filtered_df.isna().sum().sum()

print("Number of NaN values in filtered_df:", nan_count)
```
